```python
import numpy as np

#units consumed
X=np.array([50,120,180,250,320,380]).reshape(-1,1)
#corresponding electricity bill
y=np.array([500,1000,1000,1800,2500,2500])
```

```python
from sklearn.tree import DecisionTreeRegressor

reg=DecisionTreeRegressor(max_depth=3,random_state=42)
reg.fit(X,y)
```

```python
import matplotlib.pyplot as plt

X_grid=np.linspace(0,400,200).reshape(-1,1)
y_pred=reg.predict(X_grid)
plt.scatter(X,y,color="red")
plt.plot(X_grid,y_pred,color="blue")
plt.xlabel("Uniits consumed")
plt.ylabel("Electricity bill($)")
plt.title("Decision Tree Regression ")
plt.show()
```

```python
units=float(input("Enter electricity units : "))
bill=reg.predict([[units]])
print(f"Predicted electricity bill : {bill[0]:.0f}")
```

Predict salary based on experience using Decision tree regression

Input -> years of experience

Output -> salary

```python
X1=np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)
y1=np.array([3,3.5,4,4.2,6,6.5,7,7.2])
```

```python
from sklearn.model_selection import train_test_split

X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,test_size=0.2,random_state=42)
```

```python
reg1=DecisionTreeRegressor(random_state=42)
reg1.fit(X1_train,y1_train)
```

```python
X1_grid=np.linspace(1,8,100).reshape(-1,1)
y1_pred=reg1.predict(X1_grid)

plt.scatter(X1,y1,color="blue")
plt.plot(X1_grid,y1_pred,color="red")
plt.title('Decision Tree Regression')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
dt_deep=DecisionTreeRegressor(max_depth=None)
dt_deep.fit(X1_train,y1_train)

y_pred_deep=dt_deep.predict(X1_grid)
plt.scatter(X1,y1,color="blue")
plt.plot(X1_grid,y_pred_deep,color="green")
plt.title('Overfitting Tree')
plt.show()
```

```python
dt_controlled=DecisionTreeRegressor(max_depth=3)
dt_controlled.fit(X1_train,y1_train)

y_pred_controlled=dt_controlled.predict(X1_grid)
plt.scatter(X1,y1,color="blue")
plt.plot(X1_grid,y_pred_controlled,color="orange")
plt.title('Controlled Tree(max depth=3)')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score

y_test_pred_deep=dt_deep.predict(X1_test)
y_test_pred_controlled=dt_controlled.predict(X1_test)

#metrics
print("Deep Tree metrics-")
print("MSE : ",mean_squared_error(y1_test,y_test_pred_deep))
print("R2 Score : ",r2_score(y1_test,y_test_pred_deep))

print("\nControlled Tree metrics-")
print("MSE : ",mean_squared_error(y1_test,y_test_pred_controlled))
print("R2 Score : ",r2_score(y1_test,y_test_pred_controlled))
```
